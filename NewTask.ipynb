```python
class Student:                                   #Answer 1
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(self.name, self.marks)

s1 = Student("Aniket", 85)
s1.display()
```

```python
class Rectangle:                                  #Answer3
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

r = Rectangle(5, 3)
print(r.area())
```

```python
class Person:                                   #Answer7
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, dept):
        super().__init__(name, salary)
        self.dept = dept

m = Manager("Aniket", 50000, "IT")
print(m.name, m.salary, m.dept)
```

```python
class Father:                                #Answer8
    def showFather(self):
        print("Father")

class Mother:
    def showMother(self):
        print("Mother")

class Child(Father, Mother):
    pass

c = Child()
c.showFather()
c.showMother()
```

```python
class BankAccount:                             #Answer9
    def withdraw(self, amount):
        print("Withdraw:", amount)

class SavingsAccount(BankAccount):
    def withdraw(self, amount):
        if amount < 500:
            print("Minimum balance required")
        else:
            print("Withdraw:", amount)

s = SavingsAccount()
s.withdraw(300)
```

```python
class Account:                                #Answer10
    def __init__(self):
        self.__balance = 0

    def deposit(self, amt):
        self.__balance += amt

    def withdraw(self, amt):
        self.__balance -= amt

    def show(self):
        print(self.__balance)

a = Account()
a.deposit(1000)
a.withdraw(200)
a.show()
```

```python
class Circle:                                  #Answer11
    def area(self):
        return 3.14 * 2 * 2

class Square:
    def area(self):
        return 4 * 4

shapes = [Circle(), Square()]

for s in shapes:
    print(s.area())
```

```python
from abc import ABC, abstractmethod                        #ANswer12

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Bike(Vehicle):
    def start(self):
        print("Bike starts")

class Car(Vehicle):
    def start(self):
        print("Car starts")

Bike().start()
Car().start()
```

```python
class InsufficientBalanceError(Exception):              #Answer17
    pass

balance = 1000

try:
    amt = int(input("Enter withdrawal amount: "))
    
    if amt > balance:
        raise InsufficientBalanceError("Not enough balance")
    else:
        balance -= amt
        print("Transaction successful")
        print("Remaining balance:", balance)

except InsufficientBalanceError as e:
    print(e)

except ValueError:
    print("Invalid input")
```
